**EXERCÍCIO MEDIDAS DE POSIÇÃO**

```
# Importando o pandas

import pandas as pd

# Importando arquivo csv
caminho_arquivo = '/content/drive/My Drive/Aula_Data_Exploratory/DNSE2017.csv'

# Lendo o arquivo 
sinasc = pd.read_csv(caminho_arquivo, dtype=str)

campos_float = ['PESO']
campos_int = ['IDADEMAE', 'IDADEPAI', 'QTDFILVIVO', 'QTDFILMORT', 'APGAR1', 'APGAR5', 'SEMAGESTAC', 'CONSPRENAT', 'SERIESCMAE', 'QTDPARTCES', 'QTDPARTNOR', 'QTDGESTANT']

def transforma_dados(df, campos_float, campos_int):
  for c in campos_float:
    df[c] = pd.to_numeric(df[c])
  for c in campos_int:
    df[c] = pd.to_numeric(df[c]).astype(pd.Int64Dtype())
  return df
  
sinasc = transforma_dados(sinasc, campos_float, campos_int)

sinasc
```

1 ) Média de idade da mãe, peso, apgar 1 min, apgar 5 min, quantidade de filhos vivos e quantidade de filhos mortos.


```
colunas_medias = sinasc[['PESO','IDADEMAE','QTDFILVIVO','QTDFILMORT','APGAR1','APGAR5']]
medias = colunas_medias.mean().to_frame()
medias
```

2) Idade tal que metade das idades das mães são menores que elas


```
coluna_mediana = sinasc[['IDADEMAE']]
mediana_mae = coluna_mediana.median().to_frame()
mediana_mae
```

3) Apgar de 1 min que mais aparece

```
apgar1_moda = sinasc[['APGAR1']]
modas_apgar1 = apgar1_moda.mode()
modas_apgar1
```

4) Apgar de 5 min que mais aparece

```
apgar5_moda = sinasc[['APGAR5']]
modas_apgar5 = apgar5_moda.mode()
modas_apgar5
```

5)Peso da criança tal que 10% dos pesos são menores (P10)

```
quartil_10porc = sinasc[['PESO']]
peso_10porc = quartil_10porc.quantile([0.1])
peso_10porc
```

6) Peso da criança tal que 75% dos pesos são menores (Q3)


```
quartil_75porc = sinasc[['PESO']]
peso_75porc = quartil_75porc.quantile([0.75])
peso_75porc
```

7) Quantidade de filhos mortos que mais aparece

```
qt_filmort = sinasc[['QTDFILMORT']]
filhos_mortos = qt_filmort.mode()
filhos_mortos
```

8) Média de idade das mães solteiras

```
solteiras = sinasc[sinasc['ESTCIVMAE'] == '1']
solteiras[['IDADEMAE']].mean().to_frame()
```

9) Quantidade de filhos vivos das mães com até 3 anos de escolaridade tal que 25% dessa quantidade são menores (Q1)

```
escolaridade = sinasc[(sinasc['ESCMAE'] == '1') | (sinasc['ESCMAE'] == '2')]
escolaridade[['QTDFILVIVO']].quantile([0.25])
```

10) Moda da idade das mães para as crianças que nasceram com anomalia congênita

```
 anomalia = sinasc[sinasc['IDANOMAL'] == '1']
 anomalia[['IDADEMAE']].mode()
```
